```python
#TOP SPENDEERS
#Generate calculations to obtain: Purchase Count, Average Purchase Price, Total Purchase Value

#Create a Dataframe with unique values for SN
INFO_SN=purchasedata_df.groupby(["SN"])

#Generate statistics
SN_countITEMS= INFO_SN["Item ID"].count()
SN_totalPURCH=INFO_SN["Price"].sum()
SN_avg=SN_totalPURCH/SN_countITEMS



# Create new DATAFRAME and apply the appropriate formats
TopSpender=pd.DataFrame({"Purchase Count": SN_countITEMS,
                        "Average Purchase Price":SN_avg,
                        "Total Purchase Value":SN_totalPURCH})

#Sort info descending by "Total Purchase Value"
TopSpender=TopSpender.sort_values("Total Purchase Value", ascending=False)

#Formatos

TopSpender["Average Purchase Price"]=TopSpender["Average Purchase Price"].map("${:.2f}".format)
TopSpender["Total Purchase Value"]=TopSpender["Total Purchase Value"].map("${:.2f}".format)

TopSpender=TopSpender[["Purchase Count", "Average Purchase Price", "Total Purchase Value"]]

TopSpender.head()
```

```python
#MOST POPULAR ITEMS

#Create a Dataframe with unique values for item
ITEM_data=purchasedata_df.groupby(["Item ID","Item Name"])
PurchCount=ITEM_data["SN"].count()
ITEM_SumP=ITEM_data["Price"].sum()
ITEM_Price= (ITEM_SumP/PurchCount)
TotPurchVal=(ITEM_Price*PurchCount)


# Create new DATAFRAME and apply the appropriate formats
Popular=pd.DataFrame({"Purchase Count": PurchCount,
                          "Item Price": ITEM_Price,
                          "Total Purchase Value": TotPurchVal})


#Sort info descending by "Purhcase COunt"

Popular = Popular.sort_values("Purchase Count", ascending=False) 
Popular["Item Price"] = Popular["Item Price"].map("${:.2f}".format)
Popular["Total Purchase Value"] = Popular["Total Purchase Value"].map("${:.2f}".format)
Popular = Popular[["Purchase Count", "Item Price", "Total Purchase Value"]]
Popular.head(10)
```
